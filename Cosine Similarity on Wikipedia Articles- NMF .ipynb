```python
import pandas as pd
```

```python
df = pd.read_csv('wikipedia-vectors.csv', index_col=0)
df.T.head()
```

```python
titles = list(df.columns)
vocabulary=open('wikipedia-vocabulary-utf8.txt','r').read()
vocabulary=vocabulary.split('\n')
data=pd.DataFrame(df.T.values,columns=vocabulary,index=titles)
data.head()
```

# NMF

```python
from sklearn.decomposition import NMF

model = NMF(n_components=6)

model.fit(data)

nmf_features = model.transform(df.T) # Transform the articles: nmf_features

components_df = pd.DataFrame(model.components_,columns=vocabulary)

print(components_df.shape)

component = components_df.iloc[0]

print(component.nlargest(n=5))
```

```python
nmf_df=pd.DataFrame(nmf_features,index=titles)
nmf_df.head()
```
