```python
import pandas as pd
from scipy.sparse import csr_matrix

df = pd.read_csv('wikipedia-vectors.csv', index_col=0)
```

```python
df
```

```python
articles = csr_matrix(df.transpose())
titles = list(df.columns)
```

```python
vocabulary=open('wikipedia-vocabulary-utf8.txt','r').read()
vocabulary=vocabulary.split('\n')
```

```python
data=pd.DataFrame(articles.toarray(),columns=vocabulary,index=titles)
data
```

```python
# Import NMF
from sklearn.decomposition import NMF

# Create an NMF instance: model
model = NMF(n_components=6)

# Fit the model to articles
model.fit(data)

# Transform the articles: nmf_features
nmf_features = model.transform(data)

# Print the NMF features
print(nmf_features.round(2))
```

```python
# Import pandas
import pandas as pd

# Create a pandas DataFrame: df
df = pd.DataFrame(nmf_features,index=titles)

# Print the row for 'Anne Hathaway'
print(df.loc['Anne Hathaway'])

# Print the row for 'Denzel Washington'
print(df.loc['Denzel Washington'])
```

```python
df
```

```python
import pandas as pd


components_df = pd.DataFrame(model.components_,columns=vocabulary)


print(components_df.shape)


component = components_df.iloc[0]

print(component.nlargest())
```
